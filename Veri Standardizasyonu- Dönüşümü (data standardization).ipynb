```python
import numpy as np
import pandas as pd

V1=np.array([1,3,6,5,7])
V2=np.array([7,7,5,8,12])
V3=np.array([6,12,5,6,14])

df=pd.DataFrame(
    {
        "V1": V1,
        "V2": V2,
        "V3": V3}
)

df=df.astype("float")
df
```

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df) 
#ortalaması 0 std sapması 1 olacak şekilde standartlaştırma işlemi
```

NORMALİZASYON

```python
preprocessing.normalize(df)
#0-1 aralğında dönüştürme
```

MIN - MAX DÖNÜŞÜMÜ

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))
```

```python
scaler.fit_transform(df)
```

BİNARİZE DÖNÜŞÜM

```python
binarizer = preprocessing.Binarizer(threshold=5).fit(df)
#threshold 5 dedik. 5in üzerine 1 girdi, altına 0 girdi
```

```python
binarizer.transform(df)
```
